7. Visualize and understand the relationship between the multiple pairs of variables throughout different years
using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```python
sns.pairplot(df1,vars=['numcol', 'yieldpercol', 'totalprod', 'prodvalue','year']);
plt.title('Relationship between the multiple pairs of variables')
```

INFERENCE:
    Some of the pairs have no association, such as (yieldpercol,numcol), (prodvalue, yieldpercol), and (yieldpercol,totalprod). These pairs have a positive correlation because as the x axis increases, so does the y axis. Thus the two variables (totalprod,numcol), (prodvalue,numcol), and (totalprod,prodvalue) are all flowing in the same direction.
